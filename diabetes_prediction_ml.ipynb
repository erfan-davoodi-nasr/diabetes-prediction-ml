```python
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
x = df[['Pregnancies', 'Glucose', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = df['Outcome']
```

```python
scaler = StandardScaler()
scaler.fit(x)                   # convert our data to numbers between 0 to 1 for better prediction
standardized_data = scaler.transform(x)
x = standardized_data
y = df['Outcome']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=2)          # Dividing our data into train(80%) and test(20%)
```

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(x_train, y_train)      # training the support vector machine classifier
```

```python
x_test_prediction = classifier.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction, y_test)   # get accuracy score on the test data
print(f'the accuracy of our model is {testing_data_accuracy}')
```

```python
def get_float_input(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print('You have to just enter a number!!')

Pregnancies = get_float_input('Enter the number of your childbirth: ')
Glucose = get_float_input('Enter your Glucose: ')
Insulin = get_float_input('Enter your Insulin: ')
BMI = get_float_input('Enter your BMI: ')
DiabetesPedigreeFunction = get_float_input('Enter your Diabetes Pedigree Function: ')
Age = get_float_input('Enter your age: ')

# 2, 88, 58, 26, 16, 28.4, 0.766, 22              ,0   This person is not diabetic and i'm going to test it
client_information = (Pregnancies, Glucose, Insulin, BMI, DiabetesPedigreeFunction, Age)
client_information_array = np.asarray(client_information)                         # convert to array
client_information_array_reshape = client_information_array.reshape(1, -1)             # reshape our array
client_information_array_std = scaler.transform(client_information_array_reshape)              # standardize the input data
result = classifier.predict(client_information_array_std)

if (result[0] == 0):
  print("You are not diabetic")
else:
  print("You are diabetic")
```
